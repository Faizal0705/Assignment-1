```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://www.redbus.in/online-booking/tsrtc/?utm_source=rtchometile"
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text)
```

```python
soup
```

```python
RouteName = soup.select(".route")
```

```python
RouteName
```

```python
FirstBus = soup.select(".totalRoutes:nth-child(2)")
```

```python
FirstBus
```

```python
LastBus = soup.select(".totalRoutes~ .totalRoutes+ .totalRoutes")
```

```python
LastBus
```

```python
Price = soup.select(".fare")
```

```python
Price
```

```python
import pandas
```

```python
BusTiming = pandas.DataFrame(RouteName,columns=['Route Name'])
```

```python
BusTiming
```

```python
BusTiming ["First Bus Time"] = FirstBus
```

```python
BusTiming ["Last Bus Time"] = LastBus
```

```python
BusTiming ["Ticket Price"] = Price
```

```python
BusTiming
```
